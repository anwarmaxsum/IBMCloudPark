```python
df_test = df3[['drive-wheels','body-style','price']]
df_grp = df_test.groupby(['drive-wheels', 'body-style'],as_index=False).mean()
df_pivot =  df_grp.pivot(index='drive-wheels', columns='body-style')
print(df3[df3['drive-wheels']=='4wd']['body-style'].values)
df_pivot
```

```python
df_mp = df3[['make', 'price']]
df_mpgrp = df_mp.groupby(['make'], as_index=False).mean()
print(df_mpgrp)

plt.figure()
plt.bar(df_mpgrp['make'],df_mpgrp['price'])
plt.show()
```

```python
df_anova=df3[['make', 'price']]
df_grpanova = df_anova.groupby(['make'])
from scipy import stats
anovaresult = stats.f_oneway(df_grpanova.get_group('honda')['price'], df_grpanova.get_group('subaru')['price'])
print(anovaresult)
anovaresult = stats.f_oneway(df_grpanova.get_group('honda')['price'], df_grpanova.get_group('jaguar')['price'])
print(anovaresult)
```
